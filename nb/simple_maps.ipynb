```python
data_hipp = np.loadtxt("../data/mini_hip2.dat")
data_hipp[:,1] = np.rad2deg(data_hipp[:,1])
data_hipp[:,2] = np.rad2deg(data_hipp[:,2])
```

```python
plt.hist(data_hipp[:,3], log=True)
```

```python
plt.figure(figsize=(7,10))
ii = data_hipp[:,3]<6


plt.subplot(3,1,1)
plt.scatter(data_hipp[ii,1], data_hipp[ii,2], s=50*np.exp(-data_hipp[ii,3]))
#plt.axis("equal")
plt.ylim(-45,30)
plt.xlim(240,360)
plt.gca().invert_xaxis()

plt.subplot(3,1,2)
plt.scatter(data_hipp[ii,1], data_hipp[ii,2], s=50*np.exp(-data_hipp[ii,3]))
#plt.axis("equal")
plt.ylim(-30,30)
plt.xlim(120,240)
plt.gca().invert_xaxis()

plt.subplot(3,1,3)
plt.scatter(data_hipp[ii,1], data_hipp[ii,2], s=50*np.exp(-data_hipp[ii,3]))
#plt.axis("equal")
plt.ylim(-30,30)
plt.xlim(0,120)
plt.gca().invert_xaxis()
```
